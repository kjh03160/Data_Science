서류 전형 합격 여부
-----
토익 시험은 LC(듣기) 파트와 RC(독해) 파트로 이루어져 있습니다. 각 파트가 495점 만점, 총 990점이 만점입니다.


“서울 항공”에 입사하기 위해서는 토익 점수를 제출해야 하는데요. 각 파트가 최소 250점, 총 점수가 최소 600점이 되어야 서류 전형을 합격할 수 있습니다.


기존 DataFrame에 “합격 여부”라는 column을 추가하고, 합격한 지원자는 불린 값 True, 불합격한 지원자는 불린 값 False를 넣어주면 됩니다.



````python
import pandas as pd
    
df = pd.read_csv('data/toeic.csv')

cond = (df['LC'] >= 250) & (df['RC'] > 250) &(df['LC'] + df['RC'] > 600)
df['합격여부'] = cond
# 정답 출력
df
````

**```df.loc[]```** 의 사용 방식은 보통 **```df.loc[row condition, column condition]```** 입니다 (row/column 중 하나만도 가능합니다). 즉 row에 대한 조건과 column에 대한 조건이 따로 들어갑니다.

하지만 **```df.loc[:, 'B':'E'] < 80```** 의 결과값은 boolean 값이 들어가 있는 dataframe이기 때문에 (row와 column에 대한 조건이 아닌 테이블의 각 원소에 대한 조건이기 때문에) **```df.loc[df.loc[:, 'B':'E'] < 80]```** 은 허용되지 않는 것 같습니다.
