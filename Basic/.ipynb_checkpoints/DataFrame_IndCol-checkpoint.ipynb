# Index/column 설정하기
-----

* #### ```DataFrame.rename(columns={'기존 값' : '변경 값', }, inplace=True)```
* #### ```DataFrame.index.name = '값' ``` > 새로운 index 추가
* #### ```DataFrame.set_index('값') ``` > 기존 column을 index로 변경
    
     여기서 먼저 기존 index 값을 새로운 열로 추가하고 난 뒤 
    
     **```DataFrame['열이름'] = DataFrame.index ```** index를 변경해야 함.



````python
import pandas as pd

df = pd.read_csv('data/liverpool.csv', index_col = 0)
df
````

````python
df.rename(columns={'position' : 'Position'})
````

````python
df
````

````python
df.rename(columns={'position' : 'Position'}, inplace=True)
df
````

````python
df.index.name = 'Player Name'
df
````

````python
df.set_index('number')
````

````python
df.index
````

````python
df['Player Name'] = df.index
df
````

````python
df.set_index('number', inplace=True)
df
````

````python
df.index
````
