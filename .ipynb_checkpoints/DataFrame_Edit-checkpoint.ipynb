DataFrame에 값 쓰기
------
* ### ```df.loc['행', '열'] = 값```

````python
import pandas as pd

df = pd.read_csv('data/iphone.csv', index_col = 0)
df.loc['iPhone X', '메모리']
````

````python
df.loc['iPhone X', '메모리'] = '2.5GB'

df
````

DataFrame에 한 행/열의 값 쓰기
------
* ### ```df.loc['행'] = [값]```
* ### ```df['열'] = [값]```

````python
df.loc['iPhone X'] = ['2016-09-22', '4.7', '2GB', 'iOS 11.0', 'No']
df
````

````python
df['디스플레이'] = ['4.7 inch', '4.7 inch', '4.7 inch', '4.7 inch', '4.7 inch', '4.7 inch', '4.7 inch', ]
df
````

````python
df['Face ID'] = 'Yes'
df
````

DataFrame 여러 행/열의 값 쓰기
------
* ### ```df.loc[ ['행',] ] = [값]``` 해당 행의 값을 변경
* ### ```df[ ['열',] ] = [값]``` 해당 열의 값을 변경

````python
df[['디스플레이', 'Face ID']] = 'x'
df
````

````python
df.loc[['iPhone X', 'iPhone XS']] = 'o'
df
````

````python
df.loc['iPhone 7' : 'iPhone X'] = 'p'
df
````

조건으로 슬라이싱 후 변경
-----
* ### ```df.loc[df['열'] 조건] = 값```

````python
df = pd.read_csv('data/iphone.csv', index_col = 0)
print(df.dtypes)

df['디스플레이'] > 5
````

````python
df.loc[df['디스플레이'] > 5] = 'p'
df
````

----------
**```df[]```** 는 보통 **컬럼**에 대한 연산이고,

**```df.loc[]```** 은 **```df.loc[row, col]```** 이렇게 **row, column** 모두 추출할 수 있습니다 

(위와 같이 파라미터가 하나밖에 없는 경우 **```df.loc[row]```** 를 뜻합니다).

--------

````python
df.iloc[[1, 3], [1, 4]]
````

````python
df.iloc[[1, 3], [1, 4]] = 'k'
df
````

DataFrame게 값 추가/삭제
--------------

### 추가
-------

* #### ```df.loc['새로운 행'] = 값```
* #### ```df['새로운 열'] = 값```

````python
df = pd.read_csv('data/iphone.csv', index_col = 0)

df.loc['iPhone XR'] = ['2018-10-26', 6.1, '3GB', 'iOS 11.0', 'Yes']
df
````

````python
df['제조사'] = 'Apple'
df
````

### 삭제
-------

* #### ```df.drop('행', axis='index'/'columns, inplace=False)``` 

    #### axis는 colum이 아니라 row(행)을 삭제하고싶다라는 말
    
    #### inplace=False > 기존의 DataFrame을 건드리지 않겠다.
    
    **DataFrame 반환**
    
    
* #### ```df['열']```

df.drop('iPhone XR', axis='index', inplace=False)

````python
df
````

````python
df.drop('제조사', axis='columns', inplace=True)
df
````

### 여러 줄 삭제
------
* #### ```df.drop(['행', ], axis='index'/'columns, inplace=False)``` 


````python
df.drop(['iPhone 7', 'iPhone XS'], axis='index', inplace=False)
````
