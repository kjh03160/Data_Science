선 그래프 그리기
-----

```%matplotlib inline```
* #### ```DataFrame.plot(kind='line')``` : 선그래프 그리기 > kind='line' 이 default 

* #### ```DataFrame.plot(y=['col',])```, ```df[['col',]].plot()```:
    **여러 col에 관해서만 선그래프 보기, DF, Series 둘 다 가능**


````python
%matplotlib inline

import pandas as pd

df = pd.read_csv('../data/broadcast.csv', index_col=0)
df
````

````python
df.plot()
````

````python
df.plot(y='KBS')
````

````python
df.plot(y=['KBS', 'JTBC'])
````

````python
df[['KBS', 'JTBC']].plot()
````

````python
df['KBS']
````

````python
df['KBS'].plot()
````
