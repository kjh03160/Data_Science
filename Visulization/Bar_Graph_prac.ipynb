실리콘 밸리에서 일하는 남자 관리자 (Managers)에 대한 인종 분포를 막대 그래프로 다음과 같이 그려보세요.




```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('../data/silicon_valley_summary.csv')
cond = (df['gender'] == 'Male') & (df['job_category'] == 'Managers') & (df['race_ethnicity'] != 'All')

print(df[cond])
anwser = df[cond]
anwser.plot(kind='bar', x='race_ethnicity', y='count')
```
